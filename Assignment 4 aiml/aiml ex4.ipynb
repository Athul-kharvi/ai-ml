# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

X = range(1, 50)
Y = [value * 3 for value in X]

print("Values of X:")
print(*range(1,50)) 

print("Values of Y (thrice of X):")
print(Y)

# Plot lines and/or markers to the Axes.
plt.plot(X, Y)

# Set the x axis label of the current axis.
plt.xlabel('x - axis')

# Set the y axis label of the current axis.
plt.ylabel('y - axis')

# Set a title 
plt.title('Draw a line.')

# Display the figure.
plt.show()
```

# 2. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
# test.txt
# 1 2
# 2 4
# 3 1

```python
import matplotlib.pyplot as plt

# Read data from the text file
with open('test.txt', 'r') as file:
    data = [line.split() for line in file]

# Extract x and y values
x = [float(row[0]) for row in data]
y = [float(row[1]) for row in data]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the line
ax.plot(x, y, label='Line Plot')

# Set labels for the x and y axes
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# Set the title of the plot
ax.set_title('Line Plot from Data in test.txt')

# Display the legend (if multiple lines are plotted)
ax.legend()

# Show the plot
plt.show()
```

# 3. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Date,Open,High,Low,Close
# 10-03-16,774.25,776.065002,769.5,772.559998
# 10-04-16,776.030029,778.710022,772.890015,776.429993
# 10-05-16,779.309998,782.070007,775.650024,776.469971
# 10-06-16,779,780.47998,775.539978,776.859985
# 10-07-16,779.659973,779.659973,770.75,775.080017

```python
import matplotlib.pyplot as plt
import pandas as pd

# Financial data for Alphabet Inc.
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779.0, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data as lines
for column in df.columns[1:]:
    ax.plot(df['Date'], df[column], label=column)

# Set labels for the x and y axes
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Set the title of the plot
ax.set_title('Alphabet Inc. Stock Prices (Oct 3 - Oct 7, 2016)')

# Display the legend
ax.legend()

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# 5. Write a Pandas program to create and display a one-dimensional array-like object(series) containing an array of data using Pandas module. Accept the values for the series from the keyboard. Display the content.

```python
import pandas as pd

# Accept values from the keyboard as a list
data = input("Enter values separated by spaces: ").split()
try:
    data = [float(value) for value in data]  # Convert input to float
    series = pd.Series(data)

    # Display the content of the Series
    print("Series content:")
    print(series)
except ValueError:
    print("Invalid input. Please enter numeric values separated by spaces.")
```

# 7. Write a Pandas program to add, subtract, multiple and divide two Pandas Series. Accept data through the keyboard. Display the resultant series. Hint: Use +, -, *, /  operator.
# Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9].

```python
import pandas as pd

# Accept data from the keyboard
input_data1 = input("Enter values for the first series (separated by commas): ")
input_data2 = input("Enter values for the second series (separated by commas): ")

# Convert the input strings into lists of integers
series1 = list(map(int, input_data1.split(',')))
series2 = list(map(int, input_data2.split(',')))

# Create Pandas Series from the input data
ds1 = pd.Series(series1)
ds2 = pd.Series(series2)

# Perform operations
print("Add two Series:")
ds = ds1 + ds2
print(ds)

print("Subtract two Series:")
ds = ds1 - ds2
print(ds)

print("Multiply two Series:")
ds = ds1 * ds2
print(ds)

print("Divide Series1 by Series2:")
ds = ds1 / ds2
print(ds)
```

# 8. Write a Pandas program to compare the elements of the two Pandas Series for which data has been accepted through the keyboard. . Hint: Use ==, >, < operators
# Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]


```python
import pandas as pd

# Accept data from the keyboard
input_data1 = input("Enter values for the first series (separated by commas): ")
input_data2 = input("Enter values for the second series (separated by commas): ")

# Convert the input strings into lists of integers
series1 = list(map(int, input_data1.split(',')))
series2 = list(map(int, input_data2.split(',')))

# Create Pandas Series from the input data
ds1 = pd.Series(series1)
ds2 = pd.Series(series2)

# Compare the elements of the two Series
print("Comparison Results:")
print("Elements in Series 1 equal to Elements in Series 2:")
print(ds1 == ds2)

print("Elements in Series 1 greater than Elements in Series 2:")
print(ds1 > ds2)

print("Elements in Series 1 less than Elements in Series 2:")
print(ds1 < ds2)
```

# 9. Write a Pandas program to convert a dictionary(with elements empno,ename and basic) for which you accept values through the keyboard, to a Pandas series.
# Original dictionary:
# eg: {'empno': 101, 'ename': ann, 'basic': 3000, }

```python
import pandas as pd

# Accept values for the dictionary through the keyboard
empno = int(input("Enter empno: "))
ename = input("Enter ename: ")
basic = float(input("Enter basic: "))

# Create a dictionary from the user input
employee_dict = {'empno': empno, 'ename': ename, 'basic': basic}

# Convert the dictionary to a Pandas Series
employee_series = pd.Series(employee_dict)

# Display the Pandas Series
print("Pandas Series:")
print(employee_series)
```

# 10. Write a Pandas program to accept 5 elements into a numpy array and convert the NumPy array to a Pandas series.

```python
import numpy as np
import pandas as pd

# Accept elements into a NumPy array through the keyboard
input_data = input("Enter elements for the NumPy array (separated by commas): ")
elements = list(map(float, input_data.split(',')))

# Create a NumPy array from the input data
numpy_array = np.array(elements)

# Convert the NumPy array to a Pandas Series
pandas_series = pd.Series(numpy_array)

# Display the Pandas Series
print("Pandas Series:")
print(pandas_series)
```
